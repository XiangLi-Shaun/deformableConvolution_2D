```python
import os
os.environ['CUDA_VISIBLE_DEVICES'] = '1'
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### exper 2b

```python
%run main.py --action=test
```

```python
loss=np.load("/home/mzhang/dcn_2/record/validate_loss.npy")
acc=np.load("/home/mzhang/dcn_2/record/validate_accuracy.npy")
miou=np.load("/home/mzhang/dcn_2/record/validate_m_iou.npy")
start=2001


fig=plt.figure(figsize=(18,6))
epoch=range(start,start+len(loss)*100,100)

ax1=fig.add_subplot(1,3,1)
plt.plot(epoch,loss,label="loss")
plt.legend(loc="best")

ax1=fig.add_subplot(1,3,2)
plt.plot(epoch,acc,label="acc")
plt.legend(loc="best")

ax1=fig.add_subplot(1,3,3)
plt.plot(epoch,miou,label="miou")
plt.legend(loc="best")

plt.show()

print(start+100*(np.argmax(acc)))
print(max(acc))
```
